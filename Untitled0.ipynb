```
def avtomat(fieldd):  # функция автомата
    state = 'wait X'  # начально состояние
    while True:
        drawField(fieldd)  # рисовка поля
        print(state)
        # Меняет статус автомат взависимости от действия, после смены
        # на лругой статус применяються другие действия до Победного
        if state == 'wait X':
            grab(fieldd, 'X')
            state = 'move X'

        elif state == 'wait 0':
            grab(fieldd, '0')
            state = 'move 0'

        elif state == 'move X':
            state = 'wait 0'
            state = 'end X'

        elif state == 'move 0':
            state = 'wait X'
            state = 'end 0'
        elif state == 'end X':
            if checkWin(fieldd):
                state = 'win X'
            else:
                state = 'wait 0'

        elif state == 'end 0':
            if checkWin(fieldd):
                state = 'win 0'
            else:
                state = 'wait X'
        elif state == 'win X':
            print('X is winner')
            break
        elif state == 'win 0':
            print('0 is winner')
            break


def drawField(fieldd):  #
    print("—" * 13)  # рамка
    for i in range(3):
        print("|", fieldd[0 + i * 3], "|", fieldd[1 + i * 3], "|", fieldd[2 + i * 3], "|")
        print("—" * 13)  # для красивой границы


def grab(fieldd, badge):  # принимает Х или O
    valid = False
    while not valid:
        step = int(input("In which cell to put " + badge + "? "))
        valid = check_pass(fieldd, step)
        if valid:
            fieldd[step - 1] = badge
            valid = True


def check_pass(fieldd, number):  # проверяет введённые данные
    if 9 >= number >= 1:
        if str(fieldd[number - 1]) not in "X0":
            return True
        else:
            print("Wrong field")
    else:
        print('It is not a number')
        return False


def checkWin(fieldd):  # функция проверки на победу
    win = ((0, 1, 2), (3, 4, 5), (6, 7, 8), (0, 3, 6), (1, 4, 7), (2, 5, 8), (0, 4, 8), (2, 4, 6))
    for i in win:
        if fieldd[i[0]] == fieldd[i[1]] == fieldd[i[2]]:
            return fieldd[i[0]]  # кто выиграл( Х или О )
    return False


print("*_*_*_*_*_   Tic-tac-toc    _*_*_*_*_* ", )

field = [1, 2, 3, 4, 5, 6, 7, 8, 9]  # список из 9 элементов(1-9) наше поле
avtomat(field)  # вызов автомата
drawField(field)  # конечное поле
print('The result of your game')
```
